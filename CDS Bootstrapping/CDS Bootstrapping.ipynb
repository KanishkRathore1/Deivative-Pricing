```python
import pandas as pd
import numpy as np
import math as mt

#Specification of CDS Curve
spreads = pd.DataFrame({'Tenor': [0,1,2,3,4,5], 'CDS Spread': [0,0.006896,0.0077,0.0094,0.01095,0.0125], 'DF': [1,.98,.94,.92,.89,.86]})

YF = spreads.Tenor.diff().fillna(0)
spreads.insert(3,'YearFrac',YF)
recoveryrate = .4
lgd = 1-recoveryrate

survivalprob = np.zeros(len(spreads.Tenor))

k = np.zeros(len(spreads.Tenor))

for n in range (0,len(spreads.Tenor)):  
    survivalprob[0] = '1'
    survivalprob[1] = lgd/(lgd+((spreads.iloc[1,0]-spreads.iloc[0,0])*spreads.iloc[1,1]))
    if n>1:
        g = 0
        for i in range (1,n):
            num =  spreads.iloc[i,2]*(lgd*survivalprob[i-1]-((lgd+((spreads.iloc[1,3])*spreads.iloc[n,1]))*survivalprob[i]))
            g = num + g
        den = (spreads.iloc[n,2]*((lgd)+(spreads.iloc[1,3]*spreads.iloc[n,1])))
        term2 = ((survivalprob[n-1]*lgd)/((lgd)+((spreads.iloc[1,3])*spreads.iloc[n,1])))
        survivalprob[n] = (g/den)+term2
        
spreads.insert(4,'SurvivalProbability',np.around(survivalprob,4))

spreads
```

```python
#Appending hazard rates to the Probability table

hr = np.zeros(len(spreads.Tenor))
for i in range (1,len(spreads.Tenor)):
    hr[i] = -mt.log(spreads.iloc[i,4]/spreads.iloc[i-1,4])/(spreads.iloc[i,3])
spreads.insert(5,'Harard Rates',hr)
spreads
```
